```python
import tensorflow as tf
```

One Dimensional Tensor



```python
onedimvar=tf.constant(4)
print(onedimvar)
```

```python
onedimvar=tf.constant([4,1,2,3,4])
onedimvar
```

Two Dimensional Tensor

```python
tensor2=tf.constant([
    [1,2,3,4],
    [2,3,4,5],
    [1,2,4,5],
    [7,4,5,5]
])
tensor2
```

Eye

```python
eye_tensor=tf.eye(
    num_rows=5,
    num_columns=5,
    batch_shape=None,
    dtype=tf.dtypes.float32
)
eye_tensor
```

```python
eye_tensor=tf.eye(
    num_rows=5,
    num_columns=None,
    batch_shape=[5,9],
    dtype=tf.dtypes.float32
)
print(eye_tensor)
```

Fill

```python
fill_tensor=tf.fill(
    [3,3,3,3,3],11,name=None
)
print(fill_tensor)
```

One

```python
one_tensor=tf.ones(
    [5,5,15],
    dtype=tf.float32,
    name=None
)
print(one_tensor)
```

```python
ones_like_tensor=tf.ones_like(fill_tensor,
                              dtype=None,
                              name=None

)
print(ones_like_tensor)
```
